```python
import pandas as pd
```

```python
trainProcessed = pd.read_csv("./processedData/trainProcessed.csv", )
```

```python
trainProcessed[trainProcessed['Phrase'].notnull()]
```

```python
from sklearn.decomposition import TruncatedSVD
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import Normalizer
from sklearn import metrics
```

```python
aux = trainProcessed[trainProcessed['Phrase'].notnull()]
```

```python
aux.head(10)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
phrases = list(aux["Phrase"])
print(type(phrases), "\n", len(phrases), "\n", type(phrases[0]))
```

```python
vectorizer=CountVectorizer(min_df=1)

dtmTF = vectorizer.fit_transform(phrases)
df_tf = pd.DataFrame(dtmTF.toarray(), index=range(0, dtmTF.shape[0]), columns=vectorizer.get_feature_names())

df_tf.head(10)
```

```python
tfidf_transformer = TfidfTransformer()
dtmTF_IDF = tfidf_transformer.fit_transform(dtmTF)

df_tf_idf = pd.DataFrame(dtmTF_IDF.toarray(), index=range(0, dtmTF_IDF.shape[0]), columns=vectorizer.get_feature_names())
df_tf_idf.head(10)
```

```python
lsa = TruncatedSVD(25, algorithm='randomized')
dtm_lsa_tf = lsa.fit_transform(dtmTF)
dtm_lsa_tf = Normalizer(copy=False).fit_transform(dtm_lsa_tf)

lsa = TruncatedSVD(25, algorithm='randomized')
dtm_lsa_tf_idf = lsa.fit_transform(dtmTF_IDF)
dtm_lsa_tf_idf = Normalizer(copy=False).fit_transform(dtm_lsa_tf_idf)

print(dtm_lsa_tf.shape)
print(dtm_lsa_tf_idf.shape)
```

```python
df_lsa_tf = pd.DataFrame(dtm_lsa_tf, index=range(0, dtm_lsa_tf.shape[0]))
df_lsa_tf.head(10)
```

```python
df_lsa_tf_idf = pd.DataFrame(dtm_lsa_tf_idf, index=range(0, dtm_lsa_tf_idf.shape[0]))
df_lsa_tf_idf.head(10)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
y = aux.loc[:, "Sentiment"].astype("category")
y
```

```python
x_tf_train, x_tf_test, y_tf_train, y_tf_test = train_test_split(df_lsa_tf, y, random_state=27, train_size=.80)
x_tf_idf_train, x_tf_idf_test, y_tf_idf_train, y_tf_idf_test = train_test_split(df_lsa_tf_idf, y, random_state=27, train_size=.80)
```

```python
regTF_train = DecisionTreeClassifier(criterion='entropy', random_state = 27)
regTF_IDF_train = DecisionTreeClassifier(criterion='entropy', random_state = 27)

regTF_train.fit(x_tf_train, y_tf_train)
regTF_IDF_train.fit(x_tf_idf_train, y_tf_idf_train)
```

```python
y_tf_test = pd.DataFrame(y_tf_test)
y_tf_idf_test = pd.DataFrame(y_tf_idf_test)
```

```python
y_tf_test
```

```python
y_tf_test.insert(1, "Predict", list(regTF_train.predict(x_tf_test)))
y_tf_idf_test.insert(1, "Predict", list(regTF_IDF_train.predict(x_tf_idf_test)))
```

```python
y_tf_test
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_tf_test["Sentiment"], y_tf_test["Predict"])
```

```python
regTF_train.score(x_tf_test, y_tf_test["Sentiment"])
```

```python
confusion_matrix(y_tf_idf_test["Sentiment"], y_tf_idf_test["Predict"])
```

```python
regTF_IDF_train.score(x_tf_idf_test, y_tf_idf_test["Sentiment"])
```
